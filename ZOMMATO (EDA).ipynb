```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.shape
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.columns
```

```python
final_df.shape
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
#pie plot -top 3 countries that ues zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings=pd.DataFrame(final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'}))
ratings
```

Observation

1.When Rating is between 4.5 to 4.9---> Excellent

2.When Rating are between 4.0 to 3.4--->very good

3.when Rating is between 3.5 to 3.9----> good

3.when Rating is between 3.0 to 3.4----> average

4.when Rating is between 2.5 to 2.9----> average

5.when Rating is between 2.0 to 2.4----> Poo


```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating text',palette=['blue','Red','Orange','Yellow','Green','green'],data=ratings)
```

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','Green','green'])
```
